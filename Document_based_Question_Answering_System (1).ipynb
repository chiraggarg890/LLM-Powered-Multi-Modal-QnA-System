```python
import os

path =os.path.join(os.getcwd(),'Document_qna_data')
ls= []
for root, dirs, files in os.walk(path):
	for file in files:
		if(file.endswith(".m4a") or file.endswith(".mp3") or file.endswith(".webm") or file.endswith(".mp4") or file.endswith(".mpga") or file.endswith(".wav") or file.endswith(".mpeg")):
		  ls.append(os.path.join(root,file))
```

```python
length=len(ls)
#print(length)
```

```python
print(ls[0])
```

```python
if(length!=0):
  output_path=os.path.join(os.getcwd(),'Audio_to_Text')
  for inpath in ls:
    name_of_file=inpath.split('/')[-1]
    outpath=os.path.join(output_path,name_of_file)
    shutil.move(inpath,outpath )
```

```python
if(length!=0):
  all_output_paths=[]
  input_path=os.path.join(os.getcwd(),'Audio_to_Text')
  output_directory_path=os.path.join(os.getcwd(),'All_Formatted_Input_Audios')
  for filename in os.listdir('Audio_to_Text'):

    # assign files
    input_file = os.path.join(input_path,filename)
    #print(type(input_file))
    format_of_input_file=filename.split(".")[-1]
    #print(type(format_of_input_file))
    if(format_of_input_file=='m4a'):
      output_file_name=filename.split(".")[0]+".wav"
      output_file = os.path.join(output_directory_path,output_file_name)
      all_output_paths.append(output_file)
      print(output_file_name)

      # convert mp3 file to wav file
      sound = AudioSegment.from_file(input_file, format=format_of_input_file)
      sound.export(output_file, format='wav')
      os.remove(input_file)
    else:
      output_file = os.path.join(output_directory_path,filename)
      all_output_paths.append(output_file)
      shutil.copy(input_file, output_file)
      os.remove(input_file)
```

```python
def translate(audio):

    options = dict(beam_size=5, best_of=5)
    translate_options = dict(task="translate",fp16=False, **options)
    result = model.transcribe(audio,**translate_options)
    return result


```

```python
text_audio_file_path=os.path.join(os.getcwd(),'Document_qna_data/Audio_to_texxt')
for audio_file_path in all_output_paths:
  opath=audio_file_path.split('/')[-1].split('.')[0]+".txt"
  result = translate(audio_file_path)
  str_text=str(result["text"])
  final_path=os.path.join(text_audio_file_path,opath)
  os.remove(audio_file_path)
  with open(final_path,'w') as f:
    f.write(str_text)
```

#Loading documents

```python
# drive.mount("gdrive", force_remount=True)
```

```python
#os.chdir('/gdrive/MyDrive')

directory = os.path.join(os.getcwd(),'Document_qna_data')

def load_docs(directory):
  loader = DirectoryLoader(directory)
  documents = loader.load()
  return documents

documents = load_docs(directory)
len(documents)
```

#Splitting documents

```python
def split_docs(documents, chunk_size=1000, chunk_overlap=50):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs

docs = split_docs(documents)
print(len(docs))
```

#Vector search with FAISS

```python
# Load the document, split it into chunks, embed each chunk and load it into the vector store.
db = FAISS.from_documents(docs, embeddings)
```

```python
def get_similiar_docs(query):
  embedding_vector = embeddings.embed_query(query)
  docs = db.similarity_search_by_vector(embedding_vector)
  return docs
```

#Question answering using LangChain and OpenAI LLM

```python
# model_name = "text-davinci-003"
model_name = "gpt-3.5-turbo"
#model_name = "gpt-4"
llm = OpenAI(openai_api_key=OPEN_API_KEY,model_name=model_name)

chain = load_qa_chain(llm, chain_type="stuff")

def get_answer(query):
  similar_docs = get_similiar_docs(query)
  answer = chain.run(input_documents=similar_docs, question=query)
  return answer
```
